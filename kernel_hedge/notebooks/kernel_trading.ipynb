# First Model

## Data Creation

$$X_t = X_t$$
$$Y_t = X_t + Z_t$$

with $X_t$ a 0-drift GBM and Z a 0-mean OU process.


```python
#torch.manual_seed(0)
batch_x, batch_y, timesteps, d = 200, 100, 60, 1
sigma, S0, T = 0.2, 1.0, 30./365 # Black Scholes parameters
theta, mu, std = 2, 0.0, .15


dt = T/timesteps
time_grid = torch.linspace(0.0, T, timesteps).view(1, -1, 1).expand([batch_x + batch_y, timesteps, d])

dW = np.sqrt(dt)*torch.randn(batch_x + batch_y, timesteps-1, d, 2)
W = torch.cat((torch.zeros(batch_x + batch_y, 1, d), dW[..., 0].cumsum(dim=1)), dim = 1)

X = S0*torch.exp(-0.5*(sigma**2)*time_grid + sigma*W).to(device)

Z = torch.zeros(X.shape).to(X.device.type)
for t in range(Z.shape[1]-1):
    Z[:, t+1, :] = Z[:, t, :] - theta*(mu - Z[:, t, :])*dt + std*dW[:, t, :, -1]


paths = torch.cat([X, X+Z], dim=-1)
train_paths = paths[:batch_x]
test_paths = paths[batch_x:]

plt.plot(torch.linspace(0.0, T, timesteps), train_paths[0].detach().cpu())
plt.show()
```

## Model Instantiation

```python
class ItoSigKer(KernelCompute):
    def __init__(self):
        self.Kernel_Class = ItoKer()
        
    def compute_Gram(self, X, Y, sym=False):
        return self.Kernel_Class.compute_Gram(X,Y)
    
    def eta_square(self, X, time_augmented=False, max_batch=50):
        return super().eta_square(X, time_augmented, max_batch)
```

```python
Ker = ItoSigKer()

sk_trade = SigKernelTrader(Ker,
                           device='cpu', time_augment=True,
                           dyadic_order=0)

sk_trade.pre_fit(train_paths)
```

```python
sk_trade.fit(risk_aversion=0.5, reg_type='L2', regularisation=1e-15)
sk_trade.compute_pnl(test_paths)
plt.hist(sk_trade.pnl[:,-1])
print(sk_trade.pnl[:,-1].mean())
```

# Second Model

## Data Creation

$$X_t = X_t$$
$$Y_t = X_t + Z_t$$

with $X_t$ a 0-drift GBM and Z a 0-mean OU process.


```python
def generate_data(total_days, lookback_window, gbm_params, OU_params):
    
    sigma, d = gbm_params
    theta, mu, std = OU_params
    
    time_grid = torch.linspace(0.0, total_days, total_days).unsqueeze(-1).expand([total_days, d])

    dt=1
    dW = np.sqrt(dt)*torch.randn(total_days-1, d, 2)
    W = torch.cat((torch.zeros(1, d), dW[..., 0].cumsum(dim=0)), dim = 0)

    X = torch.exp(-0.5*(sigma**2)*time_grid + sigma*W).to(device)

    Z = torch.zeros(X.shape).to(X.device.type)
    for t in range(Z.shape[0]-1):
        Z[t+1, :] = Z[t, :] - theta*(mu - Z[t, :])*dt + std*dW[t, :, -1]

    paths = torch.cat([X, X+Z], dim=-1)
    
    return paths, paths.unfold(dimension=0, size=lookback_window, step=1).swapaxes(1, 2)
```

```python
train_days, test_days = 5000, 500
lookback_window = 5
gbm_params, OU_params = [0.01, 1], [-.05, 0.0, .02]

train_path, train_data = generate_data(train_days, lookback_window, gbm_params, OU_params)
test_path, test_data = generate_data(test_days, lookback_window, gbm_params, OU_params)


fig, ax1 = plt.subplots(1,2, figsize=(16,6))
ax1[0].plot(train_path)
ax1[1].plot(test_path)
```

## Model

```python
class ItoSigKer(KernelCompute):
    def __init__(self):
        self.Kernel_Class = ItoKer()
        
    def compute_Gram(self, X, Y, sym=False, max_batch=1000):
        return self.Kernel_Class.compute_Gram(X, Y, sym=sym, max_batch=max_batch)
    
    def eta_square(self, X, time_augmented=False, max_batch=1000):
        return super().eta_square(X, time_augmented, max_batch)
```

```python
Ker = ItoSigKer()

sk_trade = SigKernelTrader(Ker,
                           device='cpu', time_augment=True,
                           dyadic_order=0)

sk_trade.pre_fit(train_data)
```

```python
sk_trade.fit(risk_aversion=.5, reg_type='L2', regularisation=3e-16)
sk_trade.compute_pnl(test_data)

total_position = sk_trade.position[:,-1,:]
total_pnl = (total_position[:-1] * test_path[-total_position.shape[0]:].diff(dim=0)).cumsum(dim=0).sum(dim=-1)+1

plt.hist(sk_trade.pnl[:,-1])
print(sk_trade.pnl[:,-1].mean())
```
